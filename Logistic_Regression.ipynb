```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

**Featue Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

**Build Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
classifier.predict(sc.transform([[20,187000]]))    #Must use the .transform() because we transformed data earlier.
```

A person of age 20 with salary $187,000 is not going to purchase. What happens if we double the age to 40 but kept salary the same?

```python
classifier.predict(sc.transform([[40,187000]]))
```

The person will purchase.

**Checking the prediction with our test data**

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))   #prints predicted values of the test data then the actual values of the test data
```

You cannot use R2 value to measure the quality of your model for classification/logical regression. Instead, we use a **confusion matrix**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
